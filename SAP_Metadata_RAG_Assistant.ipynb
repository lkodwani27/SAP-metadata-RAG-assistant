# AI-Powered Enterprise Data Discovery Assistant Using RAG

## Project Overview

In modern enterprises, large volumes of data are distributed across multiple systems, departments, and platforms.  
Although organizations maintain metadata catalogs, discovering the right dataset is often difficult for both technical and business users.

Traditional keyword-based search systems:
- Lack semantic understanding
- Require technical expertise
- Do not provide conversational exploration
- Often lead to poor data discoverability

This project demonstrates how Retrieval-Augmented Generation (RAG) can transform a static metadata catalog into an intelligent conversational discovery assistant.

Using semantic search and Large Language Models (LLMs), users can explore enterprise metadata using natural language queries such as:

- “Which dataset contains customer revenue information?”
- “Find tables related to employee attendance.”
- “Which datasets are missing ownership information?”
- “Explain the customer_orders table.”

---

# Project Objectives

The primary objective of this project is to build a lightweight AI-powered metadata discovery assistant that:

- Makes enterprise metadata searchable using natural language
- Uses semantic embeddings for intelligent dataset discovery
- Retrieves grounded schema context to reduce hallucinations
- Generates business-friendly dataset explanations using an LLM
- Identifies metadata governance gaps
